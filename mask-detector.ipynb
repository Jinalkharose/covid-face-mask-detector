```python
import cv2
import numpy as np
```

```python
from keras.models import load_model

# Load the model from a file
loaded_model = load_model('my_model.h5')
```

```python
face_clsfr = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
capture = cv2.VideoCapture(0)

labels_dict = {0: 'Mask', 1: 'No Mask'}
color_dict = {0: (0, 255, 0), 1: (0, 0, 255)}
```

```python
def detect_mask_on_face(img):
    resized = cv2.resize(img, (224, 224))
    normalized = resized / 255.0
    reshaped = normalized.reshape(1, 224, 224, 3)
    result = loaded_model.predict(reshaped)
    label = np.round(result).astype(int)[0][0]
    return label  
```

```python
while(True):

    ret,img = capture.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    label = detect_mask_on_face(img)
    faces = face_clsfr.detectMultiScale(gray)  

    for (x,y,w,h) in faces:
    
        face_img = gray[y:y+w,x:x+w]
        cv2.rectangle(img, (x,y), (x+w,y+h), color_dict[label], 2)
        cv2.rectangle(img, (x,y-40), (x+w,y), color_dict[label], -1)
        cv2.putText(img, labels_dict[label], (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,255), 2)
        
        
    cv2.imshow('Live',img)
    
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break
        
cv2.destroyAllWindows()
capture.release()
```
